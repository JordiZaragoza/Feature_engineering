```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# seleccionamos el database de data.world que nos da informacion sobre accidentes en bicicleta en North Carolina,
# que usamos en la tarea del Sprint 7

# cargamos y seleccionamos las columnas que nos interesan

crash_df = pd.read_csv('north-carolina-bicycle-crash-data_1.csv', 
                       sep=';', 
                       usecols=[2, 8, 44])
crash_df.head()
```

```python
crash_df['bikesex'].value_counts()
```

```python
# descartamos las filas que tienen valor 'Unknown' en la columna 'bikesex'

crash_df = crash_df[crash_df.bikesex != 'Unknown']
crash_df
```

```python
# comprobamos que se han descartado

crash_df['bikesex'].value_counts()
```

```python
crash_df['bikerace'].value_counts()
```

```python
# descartamos las filas que tienen valor 'Unknown/Missing' en la columna 'bikerace'

crash_df = crash_df[crash_df.bikerace != 'Unknown/Missing']
crash_df
```

```python
# comprobamos que se han descartado

crash_df['bikerace'].value_counts()
```
