```python

import numpy as np
import pandas as pd
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

```python
# seleccionamos el database de data.world que nos da informacion sobre accidentes en bicicleta en North Carolina,
# que usamos en la tarea del Sprint 7

# cargamos y seleccionamos las columnas que nos interesan

crash_df = pd.read_csv('north-carolina-bicycle-crash-data_1.csv', 
                       sep=';', 
                       usecols=[2, 8, 44, 58])
crash_df.head()
```

```python
# descartamos las filas que tienen valor 'Unknown' en la columna 'bikesex'

crash_df = crash_df[crash_df.bikesex != 'Unknown']
crash_df
```

```python
# descartamos las filas que tienen valor 'Unknown/Missing' en la columna 'bikerace'

crash_df = crash_df[crash_df.bikerace != 'Unknown/Missing']
crash_df
```

```python
# descartamos las filas que tienen valor 'Unknown' en la columna 'bikeage'

crash_df = crash_df[crash_df.bikeage != 'Unknown']
crash_df
```

```python
# cambiamos los valores '70+' por '70' de la columna 'bikeage'

crash_df['bikeage'].replace({'70+':'70'}, inplace=True)
crash_df
```

```python
# descartamos las filas que tienen valor 'Unknown' en la columna 'drvrage'

crash_df = crash_df[crash_df.drvrage != 'Unknown']
crash_df
```

```python
# cambiamos los valores '70+' por '70' de la columna 'drvrage'

crash_df['drvrage'].replace({'70+':'70'}, inplace=True)
crash_df
```
